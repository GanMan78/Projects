```python


from tkinter import*

def click(event):

	global scvalue
	text=event.widget.cget("text")
	print(text)
	if text=="=":
		if scvalue.get().isdigit():
			value=int(scvalue.get())
		else:
			try:
			    value=eval(screen.get())
			except Exception as e:
				value="SyntaxError"
		scvalue.set(value)
		screen.update()
	elif text=="C":
		scvalue.set("")
		screen.update()
	elif text=="<-":
		screen.delete(first=screen.index(INSERT)-1,last=None)
		screen.update()
	else:
		scvalue.set(scvalue.get()+text)
		screen.update()
		
# Outer frame
calci=Tk()
calci.geometry("300x400")
#calci.resizable(0,0)
calci.title("Calculator by Ganesh..!!")

# Text box
scvalue=StringVar()
scvalue.set("")
screen=Entry(calci, textvar=scvalue,bg="#C1FFC1", font="verdena 35 bold")
screen.pack(expand=TRUE, fill="both", ipadx=20,pady=10,padx=10)


# Frame and buttons
#First row of numbers
f1=Frame(calci, bg="black")
b1=Button(f1, text="9",width="2", font="lucida 0 bold",relief=GROOVE,border=0)	# to remove border we use relief and border attribute
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="8", width="2", padx=0, pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="7", padx=0, width="2", pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="C", width="2",font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="<-",padx=0, pady=0,width="2", font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)
f1.pack(expand=TRUE, fill="both")

#Second row of numbers
f1=Frame(calci, bg="black")
b1=Button(f1, text="6", width="2", font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="5",width="2", padx=0, pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="4", width="2", padx=0, pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="%", width="2", padx=0, pady=0, font="lucida 0 bold ")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="/", width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)
f1.pack(expand=TRUE, fill="both")

#Third row of numbers
f1=Frame(calci, bg="black")
b1=Button(f1,text="3", width="2", font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="2",width="2", padx=0, pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="1",width="2", padx=0, pady=0, font="lucida 0 bold",relief=GROOVE,border=0)
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="*", width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text=".",width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)
f1.pack(expand=TRUE, fill="both")

#Fourth row of numbers
f1=Frame(calci, bg="black")
b1=Button(f1, text="-",width="2", padx=0, pady=0,font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="0",width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="+",width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)

b1=Button(f1, text="=",width="2", padx=0, pady=0, font="lucida 0 bold")
b1.pack(side=LEFT, padx=1, pady=1,expand=True,fill="both")
b1.bind("<Button-1>",click)
f1.pack(expand=TRUE, fill="both")


calci.mainloop()
```
